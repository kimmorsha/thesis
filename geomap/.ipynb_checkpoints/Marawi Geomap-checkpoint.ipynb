```python
filename = "../dataset/marawi_tweets.csv"

df = pd.read_csv(filename, header=None, encoding="utf-8-sig")
df
```

```python
# Get list of days

days = df[1].unique()
days = list(sorted(days.astype(str)))
days = sorted(days, key=lambda x: datetime.strptime(x, '%m/%d/%Y'))
days
```
